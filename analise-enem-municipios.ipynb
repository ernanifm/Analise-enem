## 1. Leitura do arquivo "MICRODADOS_ENEM_2021.csv" para um dataframe

```python
microdados_enem = pd.read_csv('C:/Users/ernan/Documents/Git/DADOS/MICRODADOS_ENEM_2021.csv', encoding= 'ISO-8859-1', sep = ';')
```

```python
enem_selecionados = microdados_enem.filter(items = ['CO_MUNICIPIO_ESC', 'NO_MUNICIPIO_ESC', 'CO_UF_ESC', 'SG_UF_ESC','NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC',
       'NU_NOTA_MT', 'NU_NOTA_REDACAO'])
```

```python
enem_municipios = enem_selecionados.groupby(['CO_MUNICIPIO_ESC', 'NO_MUNICIPIO_ESC', 'CO_UF_ESC', 'SG_UF_ESC'], as_index = False).mean()
```

Nas células acima, vemos que o dataframe oriundo do enem, o código do município tem 7 dígitos, e foi classificado como doubble e no do idh, 6, classificado como inteiro. Isso ocorre porque o 7o dígito, no caso da base do Enem é um dígito verificador. Para reunir os dois dataframes, poderia utilizar o nome do município, no entanto, pareceu mais fácil transformar a coluna do código do município no dataframe enem em inteiro, depois em string, e por fim, utilizar uma list comprehension para retirar o último dígito e retransformá-lo em inteiro

```python
enem_municipios['CO_MUNICIPIO_ESC'] = enem_municipios['CO_MUNICIPIO_ESC'].astype(int)
```

```python
enem_municipios.head()
```

```python
enem_municipios['CO_MUNICIPIO_ESC'] = enem_municipios['CO_MUNICIPIO_ESC'].astype(str)
```

```python
enem_municipios.head()
```

```python
enem_municipios['CO_MUNICIPIO_ESC'] = [sub[ : -1] for sub in enem_municipios['CO_MUNICIPIO_ESC']]
```

```python
enem_municipios.head()
```

```python
enem_municipios['CO_MUNICIPIO_ESC'] = enem_municipios['CO_MUNICIPIO_ESC'].astype(int)
```

```python
enem_municipios.head()
```

```python
#Agora vamos fazer um merge nos dois dataframes.
```

```python
enem_idh = pd.merge(enem_municipios, idh_reduz, how='left', left_on=['CO_MUNICIPIO_ESC'], right_on=['Código do Município'])
```

```python
enem_idh.head()
```

```python
enem_idh.plot.scatter('NU_NOTA_MT', 'IDHM Educação', figsize=(10,10))
```

```python
enem_idh.plot.scatter('NU_NOTA_REDACAO', 'IDHM Educação', figsize=(10,10))
```

```python
enem_idh['MEDIA'] =  (enem_idh['NU_NOTA_CN'] + enem_idh['NU_NOTA_CH'] 
+ enem_idh['NU_NOTA_MT'] + enem_idh['NU_NOTA_REDACAO']+enem_idh['NU_NOTA_LC'])/5
```

```python
enem_idh.plot.scatter('MEDIA', 'IDHM Educação', figsize=(10,10))
```

Vemos uma pequena tendência de que, de acordo com o IDHM Educação, a nota média tanto de matemática, como a média geral, tenham uma relação aproximadamente linear. Já para o gráfico de redação, o mesmo não foi observado claramente.
